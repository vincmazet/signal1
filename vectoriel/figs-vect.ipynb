# Représentation vectorielle des signaux

```python
import math
import numpy as np
import matplotlib.pyplot as plt

import matplotlib as mpl
mpl.rcParams['axes.grid'] = True
mpl.rcParams['text.usetex'] = True
mpl.rcParams['font.size'] = 16
mpl.rcParams['savefig.bbox'] = 'tight'
```

```python
n = np.arange(2)
u = np.array([1, 1])
d = np.array([1, 0])

# Représentation sous forme de signaux
fig = plt.figure(figsize=(10,3)) 
plt.subplot(1,2,1)
plt.stem(n, u, linefmt='C0-', basefmt=' ', markerfmt='C0o')
plt.ylim([-.1, 1.1])
plt.xlabel("$n$")
plt.xlim([-.5, 1.5])
plt.title("Échelon $u$")
plt.subplot(1,2,2)
plt.stem(n, d, linefmt='C3-', basefmt=' ', markerfmt='C3o')
plt.ylim([-.1, 1.1])
plt.xlabel("$n$")
plt.xlim([-.5, 1.5])
plt.title("Impulsion $\delta$")
plt.savefig('ex-graphe.svg', bbox_inches='tight')

# Représentation sous forme de vecteurs
fig, ax = plt.subplots(1, 1,figsize=(4,4))
ax.arrow(0, 0, u[0], u[1], head_width=0.1, length_includes_head=True, color="tab:blue")
ax.arrow(0, 0, d[0], d[1], head_width=0.1, length_includes_head=True, color="tab:red")
ax.set_xticks(np.arange(-2, 2, 1))
ax.set_yticks(np.arange(-2, 2, 1))
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_xlabel("$e_1$")
ax.set_ylabel("$e_2$")
ax.set_axisbelow(True)
plt.savefig('ex-espace.svg')
```

```python
N = 4
n = np.arange(N)
fig = plt.figure(figsize=(15,2)) 
for k in range(N):
    plt.subplot(1,N,k+1)
    phi = np.zeros(N)
    phi[k] = 1
    plt.stem(n, phi, ".", basefmt= " ")
    plt.xlabel("$n$")
    plt.title(f"$\delta_{k}$")
    plt.xlim([-.5, 3.5])
    plt.ylim([-.5, 1.5])
    plt.xticks(np.arange(0, 4))

plt.savefig('base-canonique.svg')
```
