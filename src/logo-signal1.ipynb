# Logo

```python
import numpy as np
import svgwrite
```

```python
# Taille de l'image
W, H = 540, 190

# Nombre de points du sinus cardinal
N = 80

# Espace entre le sinus cardinal et le bord de l'image
offset = 30

# Sinus cardinal
n = np.linspace(-1, 1, N)
x = np.linspace(offset, W-offset, N)
y = np.sinc(4*n)
y = 130 - y * 100

# Liste des coordonnées du sinus cardinal
points = ""
for k in range(N):
    if k==0:
        points += "M "
    else:
        points += "L "
    points += f"{x[k]} {y[k]}"

# Initialisation du SVG
dwg = svgwrite.Drawing('logo-signal1-A.svg', size=(f'{W}px', f'{H}px'))

# Gradient de couleur
gradient = dwg.linearGradient((0, 0), (1, 0), id="gradient")
gradient.add_stop_color(offset='0%',   color='rgb(0, 131, 176)', opacity=None)
gradient.add_stop_color(offset='100%', color='rgb(0, 180, 219)', opacity=None)
dwg.defs.add(gradient)

# Rectangle avec gradient
dwg.add(
    dwg.rect(insert=(0, 0), size=(f'{W}', f'{H}'), fill="url(#gradient)")
)

# Dessin du sinus cardinal
dwg.add(
    dwg.path( d=points, stroke="white", fill="none", stroke_width=6 )
)

# Enregistrement
dwg.save()
```

```python
# Taille de l'image
W, H = 50, 50

# Nombre de points du sinus cardinal
N = 70

# Sinus cardinal
n = np.linspace(-1, 1, N)
x = np.linspace(0, W, N)
y = np.sinc(5*n)
y = 30 - y*15

# Liste des coordonnées du sinus cardinal
points = ""
for k in range(N):
    if k==0:
        points += "M "
    else:
        points += "L "
    points += f"{x[k]} {y[k]}"

# Initialisation du SVG
dwg = svgwrite.Drawing('logo-signal1-B.svg', size=(f'{W}px', f'{H}px'))

# Rectangle blanc
dwg.add( dwg.rect(insert=(0, 0), size=(f'{W}', f'{H}'), fill="white") )

# Masque
mask = dwg.mask(id="themask")
mask.add(
    dwg.circle(center=(W/2, H/2), r=.48*W, fill="white")
)
dwg.defs.add(mask)

# Cercle
dwg.add(
    dwg.circle(center=(W/2, H/2), r=.48*W, stroke="black", fill="none", stroke_width=1.5)
)

# Dessin du sinus cardinal
dwg.add(
    dwg.path(d=points, stroke="black", fill="none", stroke_width=2, mask="url(#themask)")
)

# Enregistrement
dwg.save()
```
