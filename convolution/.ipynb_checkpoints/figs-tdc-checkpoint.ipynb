# Convolution

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('../signal.mplstyle')

xmin, xmax = -5, 5 # Doivent Ãªtre impairs
```

```python
# rect(t) * delta(t)

t = np.array([xmin, -.5, -.5, .5, .5, xmax])
x = np.array([0, 0, 1, 1, 0, 0])

plt.figure(figsize=(8,2))
plt.plot(t, x)
plt.ylim([-1,2])
plt.xticks(np.arange(xmin,xmax+1))
plt.yticks(np.arange(-1,2))
plt.xlabel("$t$")
plt.title("$\mathrm{rect}(t)*\delta(t)$")
plt.savefig(f"tdc-11.svg")
plt.show()
```

```python
# rect(t) * delta(t-1)

t = np.array([xmin, .5, .5, 1.5, 1.5, xmax])
x = np.array([0, 0, 1, 1, 0, 0])

plt.figure(figsize=(8,2))
plt.plot(t, x)
plt.ylim([-1,2])
plt.xticks(np.arange(xmin,xmax+1))
plt.yticks(np.arange(-1,2))
plt.xlabel("$t$")
plt.title("$\mathrm{rect}(t)*\delta(t-1)$")
plt.savefig(f"tdc-12.svg")
plt.show()
```

```python
# rect(t) * Sha_2(t)

t = np.arange(xmin+.5, xmax, 1)
t = np.concatenate([t, t])
t = np.sort(t)

K = (xmax-xmin)//2
x = np.concatenate([[0, 1, 1, 0]*K])
plt.figure(figsize=(8,2))
plt.plot(t, x)
plt.ylim([-1,2])
plt.xticks(np.arange(xmin,xmax+1))
plt.yticks(np.arange(-1,2))
plt.xlabel("$t$")
plt.title("$\mathrm{rect}(t) * I\!I\!I_2(t)$")
plt.savefig(f"tdc-13.svg")
plt.show()
```

```python
# x*x ou x est une porte

A, T = 2, 5
t = np.arange(-4*T, 4*T, T/100)
y = np.where((-2*T<t) & (t<0), A**2*(2*T+t), 0) + np.where((0<t) & (t<2*T), A**2*(2*T-t), 0)
Amax = 2*A**2*T

plt.figure(figsize=(8,2))
plt.plot(t, y)
plt.xlim([-4*T, 4*T])
plt.ylim([-Amax*.2, Amax*1.2])
plt.xlabel("$t$")
plt.title("$(x*x)(t)$")
plt.savefig(f"tdc-3.svg")
plt.show()
```
