# Signaux élémentaires

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('../src/signal.mplstyle')
```

```python
# w(t) = 3 sin(8\pi t + \pi)

t = np.linspace(-1, 1, 200)
w = 3 * np.sin(8*np.pi*t + np.pi)
plt.figure(figsize=(8,4))
plt.plot(t, w)
plt.xticks(np.arange(-1,1.1,.2))
plt.xlabel("$t$")
plt.ylabel("$w(t)$")
plt.savefig(f"tdc-w.svg")
plt.show()
```

```python
# x(t) = 1/2 sinc(t/2)

t = np.linspace(-8, 8, 200)
x = 1/2 * np.sinc(t/2)
plt.figure(figsize=(8,4))
plt.plot(t, x)
plt.xticks(np.arange(-8,9,2))
plt.xlabel("$t$")
plt.ylabel("$x(t)$")
plt.savefig(f"tdc-x.svg")
plt.show()
```

```python
# y[n] = sin(\pi n)

n = np.arange(-5, 6)
y = np.sin(np.pi*n)
plt.figure(figsize=(8,4))
plt.stem(n, y, basefmt=" ")
plt.xticks(np.arange(-4,6,1))
plt.ylim(-1,1)
plt.xlabel("$n$")
plt.ylabel("$y[n]$")
plt.savefig(f"tdc-y.svg")
plt.show()
```

```python
# z = rect(n/4-1/2) - delta[n-2]

n = np.arange(-8, 9)
rect  = np.where( abs(n/4-1/2)<=.5, 1, 0)
delta = np.where( (n-2)==0, 1, 0)
z = rect - delta
plt.figure(figsize=(8,4))
plt.stem(n, z, basefmt=" ")
plt.xticks(np.arange(-8,9,1))
plt.xlabel("$n$")
plt.ylabel("$z[n]$")
plt.savefig(f"tdc-z.svg")
plt.show()
```
