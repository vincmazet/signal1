#Â Signaux ne respectant pas les conditions de Dirichlet

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('../signal.mplstyle')
mpl.rcParams['font.size'] = 16
```

```python
signals = [
    lambda n : n,
    lambda n : np.sin(1/(6*n)),
    lambda n : np.where( np.sin(1/(6*n))>0, 1, -1),
]
K = len(signals)

n = np.linspace(-.3,.3,3000)
n = n**3

fig, axs = plt.subplots(1, K, figsize=(13,4))
for k in range(K):
    x = signals[k]
    axs[k].plot(n,x(n))
plt.savefig("dirichlet.svg")
plt.show()
```
