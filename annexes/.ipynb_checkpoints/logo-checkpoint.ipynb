# Logo

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Initialisation de l'image
C, R = 540, 190
img = np.zeros((R,C,3), dtype="uint8")

# Grandient de couleur
c1 = [0, 131, 176]
c2 = [0, 180, 219]
for band in range(3):
    clr = np.linspace(c1[band], c2[band], C)
    img[:,:,band] = clr

# Sinus cardinal
N = 500
n = np.linspace(-1, 1, N)
x = np.sinc(n*4)
offset = (C-N)//2
for c in range(N):
    row = int(np.round( 130 - x[c]*100 ))
    col = c + offset
    img[row,col,:] = [255, 255, 255]

# Affichage
plt.imshow(img)
```
